```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("advertising.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
plt.title("TV v/s Sales")
plt.xlabel("TV")
plt.ylabel("Sales")
plt.scatter("TV" , "Sales" , data=df)
plt.show()
```

```python
plt.title("Radio v/s Sales")
plt.xlabel("Radio")
plt.ylabel("Sales")
plt.scatter("Radio" , "Sales" , data=df)
plt.show()
```

```python
plt.title("Newspaper v/s Sales")
plt.xlabel("Newspaper")
plt.ylabel("Sales")
plt.scatter("Newspaper" , "Sales" , data=df)
plt.show()
```

```python
sns.pairplot(df , x_vars = ["TV","Radio","Newspaper"] , y_vars = "Sales" , height = 4)
plt.show()
```

```python
advertising=["TV","Radio","Newspaper"]
tv=df['TV'].sum()
radio=df['Radio'].sum()
newspaper=df['Newspaper'].sum()
adv=[tv,radio,newspaper]
colors=["r","b","y"]
plt.pie(adv,labels=advertising,colors=colors,autopct="%0.2f%%",shadow=True,radius=1.4,startangle=90)
plt.show()
```

```python
sns.pairplot(df , diag_kind = "kde")
plt.show()
```

```python
sns.heatmap(df.corr() , cmap="coolwarm" , annot=True)
plt.show()
```

```python
sns.displot(df["Sales"])
plt.show()
```

```python
sns.jointplot(x="TV" , y = "Sales" , data = df , kind = "reg")
```
